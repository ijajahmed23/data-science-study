```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
print(max(x))
print(min(x))
print(max(y))
print(min(y))
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
```

```python
plt.scatter(x,y,c='r')  #c is color of the plot
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("thai is my title")
```

```python
plt.scatter(x,y,c='#0AA716') #c is color of the plot...Hexa value can also be given in c
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("thai is my title")
```

```python
plt.scatter(x,y,c='#0AA716',alpha=0.2) #alpha is blending of color in 0-1...0 transparent,,1 opaque
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("thai is my title")
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='#0AA716',alpha=0.9,marker='X') #marketr is the plot unit
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my title")
plt.grid()
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(7,3))
plt.scatter(x,y,c='r',alpha=0.8,marker='X')
plt.xlabel('Phase')
plt.ylabel('Amplitude')
plt.title('Sin curve')
plt.grid()
```

```python
plt.plot(x,y,'-r')
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Sine Wave')
plt.grid()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
