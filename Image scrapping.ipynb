```python
query="karl marx"
response=requests.get(f"https://www.google.com/search?sxsrf=AJOqlzXMJ0UOvieQAQFSpniTKhC3srK25w:1677082425821&q={query}&tbm=isch&sa=X&ved=2ahUKEwiR4v_xwqn9AhUlUPUHHQQrA2gQ0pQJegQIDBAB&biw=1536&bih=755&dpr=1.25")
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
image_tags=soup.find_all("img")
```

```python
len(image_tags)
```

```python
image_tags[0]["src"]
```

```python
del image_tags[0]
```

```python
image_data_mongo=[]
for i in image_tags:
    image_url=i["src"]
    image_data=requests.get(image_url).content
    my_dict={"index":image_url,"image":image_data}
    image_data_mongo.append(my_dict)
    with open(os.path.join(save_dir,f"{query}_{image_tags.index(i)}.jpg"),"wb") as f:
        f.write(image_data)
```
