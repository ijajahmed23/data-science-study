```python
data1=pd.DataFrame({"key1":[1,2,4,5,6],
    "key2":[4,5,6,7,8],
    "key3":[3,4,5,6,6]   
})
```

```python
data1
```

```python
data2=pd.DataFrame({"key1":[1,2,45,6,67],
    "key4":[56,5,6,7,8],
    "key5":[3,56,5,6,6]   
})
```

```python
data2
```

```python
#merge operatrion
pd.merge(data1,data2)
#inner merge operation
```

```python
pd.merge(data1,data2,how='left')
#keeps everything from left..then matches key1 in both dataset and appends key4 and key5 values from data2
```

```python
pd.merge(data1,data2,how='right')
#keeps everything from right..then matches key1 in both dataset and appends key2 and key3 values from data1
```

```python
pd.merge(data1,data2,how='outer')
#keeps the entire dataset from both data1 and data2
```

```python
pd.merge(data1,data2,how='cross')
#does a cross product..no of records..n1Xn2
```

```python
pd.merge(data1,data2,how='left',on='key1')
```

```python
pd.merge(data1,data2,how='left',left_on='key2',right_on='key4')
```

##Join

```python
data1=pd.DataFrame({'key1':[1,2,4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]},
                   index=['a','b','c','d','e'])
```

```python
data2=pd.DataFrame({'key11':[1,2,4,5,6],
                    'key22':[4,5,6,7,8],
                    'key33':[3,4,5,6,6]},
                   index=['a','b','h','i','j'])
```

```python
data1
```

```python
data2
```

```python
data1.join(data2)
#default is left join
#join matches by indexes only not columns like merge
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='cross')
```

```python
#merge will check column values 
#join will check index values
```
